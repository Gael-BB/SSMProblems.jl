```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_drift = pd.read_csv('samples_drift.csv')
df_drift
```

```python
v_max = 3
v_min = -2
bars = 100

bins = np.linspace(v_min, v_max, bars+1)
plt.figure(figsize=(8, 5), dpi=200)
plt.hist2d(df_drift.iloc[:, 0], df_drift.iloc[:, 1], bins=bins, density=True, cmap='magma')
plt.title(r'Drift ($b$) Posterior')

plt.colorbar(label='Density')
plt.xlim(v_min, v_max)
plt.ylim(v_min, v_max)
```

```python
df_scale = pd.read_csv("samples_scale.csv")
df_scale
```

```python
bins = np.linspace(0, 2, 102)
plt.figure(figsize=(8, 5), dpi=200)
plt.hist(df_scale.iloc[:, 0], bins=bins, density=True, color='skyblue', edgecolor='navy', alpha=0.7)
plt.title(r'Scale ($\sigma^2$) Posterior', fontsize=14)
plt.xlabel(r'$\sigma^2$', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
