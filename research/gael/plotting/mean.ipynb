```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("output_scale.csv")
del df['seed']
df.head()
```

```python
mean_df = df.groupby('algorithm').mean().reset_index()
std_df = df.groupby('algorithm').std().reset_index()
```

```python
metrics = ['squared_error', 'effective_sample_size', 'runtime']
std_df_pct = std_df.copy()
std_df_pct[metrics] = std_df_pct[metrics] / mean_df[metrics]
std_df_pct
```
